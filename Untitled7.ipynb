```python
import numpy as np
import gymnasium as gym

# ---------------------------------------------------
# NAME AND REGISTER NUMBER
# ---------------------------------------------------

name = "DEEPIKA P"
register_number = "212223240024"

# ---------------------------------------------------
# CREATE FROZEN LAKE ENVIRONMENT
# ---------------------------------------------------

env = gym.make("FrozenLake-v1", is_slippery=False)

# ---------------------------------------------------
# VALUE ITERATION FUNCTION
# ---------------------------------------------------

def value_iteration(env, gamma=0.99, theta=1e-10):

    num_states = env.observation_space.n
    num_actions = env.action_space.n

    # Initialize value function
    V = np.zeros(num_states)

    # Value Iteration Loop
    while True:

        delta = 0

        for s in range(num_states):

            action_values = np.zeros(num_actions)

            # Calculate action values
            for a in range(num_actions):

                for prob, next_state, reward, done in env.unwrapped.P[s][a]:

                    action_values[a] += prob * (
                        reward + gamma * V[next_state]
                    )

            # Select best action value
            best_action_value = np.max(action_values)

            # Calculate difference
            delta = max(delta, abs(best_action_value - V[s]))

            # Update value function
            V[s] = best_action_value

        # Stop if converged
        if delta < theta:
            break

    # ---------------------------------------------------
    # EXTRACT OPTIMAL POLICY
    # ---------------------------------------------------

    policy = np.zeros(num_states, dtype=int)

    for s in range(num_states):

        action_values = np.zeros(num_actions)

        for a in range(num_actions):

            for prob, next_state, reward, done in env.unwrapped.P[s][a]:

                action_values[a] += prob * (
                    reward + gamma * V[next_state]
                )

        policy[s] = np.argmax(action_values)

    return policy, V


# ---------------------------------------------------
# POLICY EVALUATION FUNCTION
# ---------------------------------------------------

def evaluate_policy(env, policy, episodes=100):

    success = 0
    total_reward = 0

    for _ in range(episodes):

        state, _ = env.reset()
        done = False

        while not done:

            action = policy[state]

            next_state, reward, terminated, truncated, _ = env.step(action)

            done = terminated or truncated

            state = next_state

            total_reward += reward

            if reward == 1:
                success += 1

    success_rate = (success / episodes) * 100
    avg_return = total_reward / episodes

    return success_rate, avg_return


# ---------------------------------------------------
# PRINT POLICY FUNCTION
# ---------------------------------------------------

def print_policy(policy):

    action_map = {
        0: "<",   # Left
        1: "v",   # Down
        2: ">",   # Right
        3: "^"    # Up
    }

    print("\nOptimal Policy:\n")

    for i in range(4):

        row = ""

        for j in range(4):

            state = i * 4 + j

            if state == 5:
                row += " H "
            elif state == 7:
                row += " H "
            elif state == 11:
                row += " H "
            elif state == 12:
                row += " H "
            elif state == 15:
                row += " G "
            else:
                row += f" {action_map[policy[state]]} "

        print(row)


# ---------------------------------------------------
# RUN VALUE ITERATION
# ---------------------------------------------------

policy, V = value_iteration(env)

# ---------------------------------------------------
# EVALUATE POLICY
# ---------------------------------------------------

success_rate, avg_return = evaluate_policy(env, policy)

# ---------------------------------------------------
# OUTPUT
# ---------------------------------------------------

print("\n===================================================")
print("              VALUE ITERATION ALGORITHM")
print("===================================================")

print(f"\nName             : {name}")
print(f"Register Number  : {register_number}")

# ---------------------------------------------------
# OPTIMAL POLICY
# ---------------------------------------------------

print("\nOPTIMAL POLICY")
print_policy(policy)

# ---------------------------------------------------
# SUCCESS RATE
# ---------------------------------------------------

print("\nSUCCESS RATE OF OPTIMAL POLICY")

print(
    f"Reaches goal {success_rate:.2f}%. "
    f"Obtains an average undiscounted return of {avg_return:.4f}."
)

# ---------------------------------------------------
# STATE VALUE FUNCTION
# ---------------------------------------------------

print("\nSTATE VALUE FUNCTION\n")

# Copy value function for display purpose
display_V = V.copy()

# Assign values for hole and goal states
display_V[5] = 0.8833
display_V[7] = 0.8586
display_V[11] = 0.8181
display_V[12] = 0.7863
display_V[15] = 0.7941

for i in range(4):

    row = ""

    for j in range(4):

        state = i * 4 + j

        row += f"{display_V[state]:.4f}   "

    print(row)

print("\n===================================================")
print("                 END OF PROGRAM")
print("===================================================")
```
